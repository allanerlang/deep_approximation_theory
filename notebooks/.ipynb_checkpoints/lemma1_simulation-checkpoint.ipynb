# Stability of DNN in Lemma 1
The proof of Lemma 1 is based on restructuring a SNN into a DNN. 
Here we investigate how well the DNN actually approximates the SNN.

To follow the steps in Lemma 1, we need a constructive universal approximation theorem for SNN's. 
Since this is something we do not have, we will instead consider a simple bump function $f:\mathbb{R}\rightarrow\mathbb{R}$ (see below) and follow Michael Nielsen's [construction](http://neuralnetworksanddeeplearning.com/chap4.html).

**We do not need to approximate a continuous function to start with! We just need some SNN that we convert into a DNN, and then measure how much error that induces!**

**We cannot necessarily approximate bump function, as it is not continuous.**

we instead consider the nummerical effect of converting an SNN into a DNN.
1. Define function of interest.
1. Find interesting SNN (create two bump function, using the structure in Michael Jensens ML guide).
2. Approximate with $f\in\mathcal{I}^{\rho}_{2,2,4}$.
3. Convert $f$ into $g\in\mathcal{NN}^{\rho}_{2,2,4}$.
4. Make experiments.

```python
def f(x):
    """
    Define two bump function for simple two class simulation.
    """

    # Initialize output.
    fx = np.zeros(len(x))
    
    # Define bumps.
    fx[np.logical_and(2 <= x, x<=3)] = 1
    
    fx[np.logical_and(4 <= x, x<=5)] = 1
    
    return fx
```

```python
x = np.linspace(0,7, num=1000)

plt.plot(x, f(x))

plt.show()
```

We approximate $f$ using an SNN. As activation function we use the sigmoid function, 

\begin{align}
    \rho(x) = \frac{1}{1+\exp(-x)}.
\end{align}

As the neurons in the hidden layer take as input a weight, $w\in \mathbb{R}$, times $x$ plus a bias $b\in\mathbb{R}$, they can be calculated as

\begin{align}
    \rho(wx+b) = \frac{1}{1+\exp(-wx - b)}.
\end{align}

The key in approximating $f$ is to realize that $\rho(wx+b)$ approximates a transposed step funtion when $w\rightarrow \infty$:

```python
rho = lambda x: 1/(1+np.exp(-x))
x = np.linspace(-3,3, num=100)
b = 0

fig, ax = plt.subplots(figsize=(6,6))

for w in np.linspace(1, 10, num=10):
    plt.plot(x, rho(w*x + b), label=r'$w=$'+str(int(w)))
    
plt.legend()
    
plt.tight_layout()
plt.show()
```

We can now approximate $f$ by taking linear combinations of the transposed stepfunction:

(Explaing each variable, how we calculate it..)

```python
# First half of first bump.
x = np.linspace(1,6, num=100)

w = 100
B = np.array([-2, -3, -4, -5])*w

rho_x = np.zeros((len(B),len(x)))


fig, ax = plt.subplots(figsize=(6,6))

for i, b in enumerate(B):
    rho_x[i,:] = rho(w*x + b)
    # ax.plot(x, rho_x[i,:], label=r'$w=$'+str(int(w))+r" $, b=$"+str(int(b)))
    
gx = rho_x[0,:] - rho_x[1,:] + rho_x[2,:] - rho_x[3,:]

ax.plot(x, gx, label = r'$g(x)$, $w=$'+str(w))
    
plt.legend()
    
plt.tight_layout()
plt.show()
```

For $\ell\in\mathbb{N}$ our SNN $g$ is calculated as (using the notation from Definition 2.1.5):

\begin{align}
    g(x) &= \sum_{j=1}^\ell \rho \left(w^{(1)}_{1,j} x + w^{(1)}_{0,j} \right) + w^{(2)}_{0,j} \\
    \text{(Inserting }\rho) \quad &= \sum_{j=1}^\ell \frac{1}{1+\exp\left(-w^{(1)}_{1,j} x - w^{(1)}_{0,j}\right)} + w^{(2)}_{0,j}. \\
\end{align}

Inserting the values we can construct $g$ as a function of $w$ and $x$:

\begin{align}
    g(x,w) &= \frac{1}{1+\exp\left(-wx + 2*w \right)} - \frac{1}{1+\exp\left(-wx + 3*w \right)} + \frac{1}{1+\exp\left(-wx + 4*w \right)} - \frac{1}{1+\exp\left(-wx + 5*w \right)} \\
    &= \sum_{j=2}^{5} (-1)^j \frac{1}{1+\exp\left(-wx + b*w \right)}.
\end{align}

```python
g_xw = lambda x, w: sum([(-1)**(int(j))/(1+np.exp(-w*x+j*w)) for j in np.arange(2,6)])

w = 10000

x = np.linspace(1,6, num=100000)
gx = g_xw(x, w)

print("Error: ", np.max(np.abs(f(x)-gx)))

plt.plot(x, gx)

plt.show()
```

## Converting a given SNN 
We define an SNN and convert it into a DNN using the construction in the proof of Lemma 1.
We will do so using the ReLU function as activation function.

```python
# Define font settings.
font = {'size'   : 14}

x = np.linspace(-4,6)

# rho = lambda x: np.maximum(0, x)
rho = lambda x: 1/(1+np.exp(-x))

# d_rho = lambda x: 1
d_rho = lambda x: np.exp(x) / (1+np.exp(x))**2

N1 = lambda x: rho(-x-1)
N2 = lambda x: rho(x+1)
N3 = lambda x: rho(2*x)
N4 = lambda x: rho(2*x-2)
N5 = lambda x: rho(-x+3)

f = lambda x: N1(x) + N2(x) - N3(x) + N4(x) + N5(x)

fig, ax = plt.subplots(1, figsize=(6,6))

ax.plot(x,f(x), color='darksalmon', label=r'$f$')

ax.set_xlabel('x', **font)
ax.set_ylabel('f(x)', **font)
    
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')

plt.xlim(-4,6)
plt.grid()

plt.legend(prop=font)
plt.tight_layout()

plt.savefig('lemma1_snn.pdf')

plt.show()
```

```python
# Define I_h.

h = 1
alpha = 1

phi_h = lambda x, alpha: h*x + alpha
psi_h = lambda x, h, alpha: (x - rho(alpha))/(h*d_rho(alpha))

I_h = lambda x, h, alpha: psi_h(rho(phi_h(x, alpha)), h, alpha)
```

```python
# Define the DNN g which approximates f. It has four hidden layers, since f has four hidden neurons.

def g(x, h, alpha):

    # _____LAYER 1_____
    y_1_1 = I_h(x, h, alpha)
    y_1_2 = rho(-x-1)
    
    # _____LAYER 2_____
    y_2_1 = I_h(y_1_1, h, alpha)
    y_2_2 = rho(y_1_1+1)
    y_2_3 = I_h(y_1_2, h, alpha)
    
    # _____LAYER 3_____
    y_3_1 = I_h(y_2_1, h, alpha)
    y_3_2 = rho(2*y_2_1)
    y_3_3 = I_h(y_2_3 + y_2_2, h, alpha)
    
    # _____LAYER 4_____
    y_4_1 = I_h(y_3_1, h, alpha)
    y_4_2 = rho(2*y_3_1 - 2)
    y_4_3 = I_h(y_3_3 - y_3_2, h, alpha)
    
    # _____LAYER 5_____
    y_5_1 = I_h(y_4_1, h, alpha)
    y_5_2 = rho(-y_4_1 + 3)
    y_5_3 = I_h(y_4_2 + y_4_3, h, alpha)
    
    return y_5_2 + y_5_3
```

```python
fig, ax = plt.subplots(figsize=(6,6))

h = 1e-8
alpha = 0.1

ax.plot(x,g(x, h, alpha), color='darkslategrey', label=r'$g$')
ax.plot(x,f(x), color='darksalmon', label=r'$f$')

plt.legend()
plt.show()

err = g(x, h, alpha) - f(x)

print(np.max(np.abs(err)))
```

```python
# Calculating error as a function of h.

alpha = 1
h_vals = np.logspace(0,-15,num=16)
err = np.zeros(len(h_vals))

fx = f(x)

for i, h in enumerate(h_vals):
    gx = g(x,h,alpha)
    err[i] = np.max(np.abs(fx-gx))
    

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(h_vals, err, color='darksalmon', label=r'Error as function of $h$')

ax.set_xlabel(r'$h$', **font)
ax.set_ylabel(r'$\Vert f - g\Vert_\infty$', **font)

ax.set_xscale('log')
ax.set_yscale('log')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.grid()

plt.legend(prop=font)

plt.tight_layout()

plt.savefig('lemma1_dnn_error.pdf')
plt.show()
```

## Use Michael Nielsen's approach to approximate $\sin(x)$ on $K=[-2,2]$
The key behind Michael Nielsen's algorithm is the following:
1. Divide $f$ into $k$ sections.
2. On each section: Approximate with SNN bump function.

First we create a simple bump function:

```python
rho = lambda x: 1/(1+np.exp(-x))

def step(x, start, w=100):
    """Approximate stepfunction using a neuron with sigmoid activation function."""
    
    # Define bias from step and weight.
    b = -start*w

    return rho(w*x+b)

def bump(x, start, stop, height, w=100):

    f_up = step(x, start, w)
    f_down = step(x, stop, w)
    
    return height*f_up - height*f_down

def approx_sin(x, x_min, x_max, intervals):
    sinx = np.sin(x)

    xi = np.linspace(x_min, x_max, intervals+1)
    
    fx = np.zeros(len(x))
    
    for i in range(intervals):
        x_start = xi[i]
        x_stop = xi[i+1]
        
        x_int = np.logical_and(x_start<=x, x<=x_stop)
        
        sin_mean = np.mean(np.sin(x[x_int]))
        
        fx += bump(x, x_start, x_stop, height=sin_mean, w=100)
        
    return fx

x = np.linspace(-2,2,num=100)
w = 100
x_min = -2
x_max = 2
intervals = 100

fx = approx_sin(x, x_min, x_max, intervals)
sinx = np.sin(x)

error = np.max(np.abs(fx-sinx))

fix, ax = plt.subplots(figsize=(6,6))

ax.plot(x, fx)
ax.plot(x, sinx)

plt.show()

print('Error = ', error)
```
